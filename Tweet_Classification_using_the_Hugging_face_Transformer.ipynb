```python
import datasets
import huggingface_hub
import matplotlib.pyplot as plt
import transformers
```

```python
# Text Classification
from transformers import pipeline
classifier = pipeline('text-classification')
```

```python
import pandas as pd
test_df=pd.read_csv('/content/test.csv')
test_df
```

```python
def load_test_data():
    test_reviews = test_df['text'].tolist()
    return test_reviews
```

```python
test_reviews = load_test_data()
test_reviews[:5]
```

```python
target = []
for review in test_reviews:
  result=classifier(review)
  if result[0]['label'] == 'POSITIVE':
    target.append(0)
  else:
    target.append(1)
```

```python
id = test_df['id'].tolist()
```

```python
import csv

# Sample lists (ensure they have the same length)
#data1 = ["apple", "banana", "orange"]
#data2 = [10, 20, 15]  # Prices or any corresponding data
column_names=['id','target']
# Open the CSV file for writing
with open('Submission.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(column_names)

  # Write each row (zip combines elements from both lists)
  for item1, item2 in zip(id,target):
    writer.writerow([item1, item2])

print("CSV file created with corresponding data!")
```
