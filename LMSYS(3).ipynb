```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Datasets/train (1).csv")
data
```

```python
selected_columns = [0,3,4,5,6,7,8]
df = data.iloc[:,selected_columns]
df
```

```python
LABEL_COLUMNS = df.columns.tolist()[4:]
LABEL_COLUMNS
```

```python
text1 = df['response_a'].tolist()
text2 = df['response_a'].tolist()
labels=[]
for i in range(len(text1)):
  data_row = df.iloc[i]
  label_row=data_row[LABEL_COLUMNS]
  labels.append(label_row)
```

```python
labels=[[float(x) for x in sublist] for sublist in labels]
text1[:5],text2[:5],labels[:5]
```

```python
from transformers import RobertaTokenizer, RobertaModel
from transformers import RobertaForSequenceClassification, AutoModelForSequenceClassification
import torch
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset
```

```python

checkpoint = "roberta-base"
tokenizer = RobertaTokenizer.from_pretrained(checkpoint)
model = RobertaForSequenceClassification.from_pretrained(checkpoint, num_labels=len(labels[0]),
                                                            problem_type="multi_label_classification")
```

```python
# Lets build custom dataset
class CustomDataset(Dataset):
  def __init__(self, text1,text2, labels, tokenizer, max_len=128):
    self.text1 = text1
    self.text2 = text2
    self.labels = labels
    self.tokenizer = tokenizer
    self.max_len = max_len

  def __len__(self):
    return len(self.text1)

  def __getitem__(self, idx):
    text1 = str(self.text1[idx])
    text2 = str(self.text2[idx])
    label = torch.tensor(self.labels[idx])

    encoding = self.tokenizer(text1,text2, truncation=True, padding="max_length", max_length=self.max_len, return_tensors='pt')

    return {
        'input_ids': encoding['input_ids'].flatten(),
        'attention_mask': encoding['attention_mask'].flatten(),
        'labels': label
    }
```

```python
train_text1, val_text1, train_labels, val_labels = train_test_split(text1, labels,
                                                                    test_size=0.2, random_state=42)
train_text2, val_text2, _, _ = train_test_split(text2, labels,test_size=0.2, random_state=42)
```

```python
train_dataset = CustomDataset(train_text1,train_text2, train_labels, tokenizer)
val_dataset = CustomDataset(val_text1,val_text2, val_labels, tokenizer)
```

```python
# Multi-Label Classification Evaluation Metrics
import numpy as np
from sklearn.metrics import roc_auc_score, f1_score, hamming_loss
from transformers import EvalPrediction
import torch


def multi_labels_metrics(predictions, labels, threshold=0.5):
  sigmoid = torch.nn.Sigmoid()
  probs = sigmoid(torch.Tensor(predictions))

  y_pred = np.zeros(probs.shape)
  y_pred[np.where(probs>=threshold)] = 1
  y_true = labels

  f1 = f1_score(y_true, y_pred, average = 'macro')
  roc_auc = roc_auc_score(y_true, y_pred, average = 'macro')
  hamming = hamming_loss(y_true, y_pred)

  metrics = {
      "roc_auc": roc_auc,
      "hamming_loss": hamming,
      "f1": f1
  }

  return metrics

def compute_metrics(p:EvalPrediction):
  preds = p.predictions[0] if isinstance(p.predictions, tuple) else p.predictions

  result = multi_labels_metrics(predictions=preds,
                                labels=p.label_ids)

  return result
```

```python
# Training Arguments
from transformers import TrainingArguments, Trainer

args = TrainingArguments(
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    output_dir = './results',
    num_train_epochs=5,
    save_steps=1000,
    save_total_limit=2,
    learning_rate=2e-5,
    evaluation_strategy="epoch"
)

trainer = Trainer(model=model,
                  args=args,
                  train_dataset=train_dataset,
                  eval_dataset = val_dataset,
                  compute_metrics=compute_metrics)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```

```python
text1 = "You have two oranges today."
text2 = "You still have three oranges. Eating an orange yesterday does not affect the number of oranges you have today."

encoding = tokenizer(text1,text2, return_tensors='pt')
encoding.to(trainer.model.device)

outputs = trainer.model(**encoding)
```
